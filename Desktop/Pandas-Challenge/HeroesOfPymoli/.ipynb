```python
#Dependencies and setup
import pandas as pd
import numpy as np

#File to Load
file_to_load = '/Users/jw/Desktop/Pandas-Challenge/purchase_data.csv'

#Read Purchasing File and store into Pandas data frame
purchase_data_DF = pd.read_csv(file_to_load)

purchase_data_DF.head()
```

PLAYER COUNT

Display Total Number of Players


```python
#Count Total Players by SN
Total_Players = purchase_data_DF['SN'].count()

#Create DataFrame
Total_Players_DF = pd.DataFrame({"Total Players": [Total_Players]})

#Display DataFrame
Total_Players_DF
```

PURCHASING ANALYSIS (TOTAL)

Run basic calculations to obtain number of unique items, average price, etc.


Create a summary data frame to hold the results

Optional: give the displayed data cleaner formatting


Display the summary data frame

```python
#Basic Calculations
Number_of_Items = len(purchase_data_DF['Item Name'].unique())

Avg_Price = purchase_data_DF['Price'].mean()

Number_of_Purchases = len(purchase_data_DF['Purchase ID'].unique())

Total_Revenue = sum(purchase_data_DF['Price'])

#Create DataFrame
Purchasing_Analysis_DF = pd.DataFrame({"Number of Unique Items": [Number_of_Items],
                                      "Average Price": [Avg_Price],
                                      "Number of Purchases": [Number_of_Purchases],
                                      "Total Revenue": [Total_Revenue]})

#Display DataFrame
Purchasing_Analysis_DF
```

GENDER DEMOGRAPHICS

Percentage and Count of Male Players


Percentage and Count of Female Players


Percentage and Count of Other / Non-Disclosed

```python
#Percentage Calculations
gender_dem_totals = purchase_data_DF["Gender"].value_counts()

gender_dem_perc = gender_dem_totals / Total_Players * 100

#Create DataFrame
gender_dem_df = pd.DataFrame({"Total Count": gender_dem_totals,
                             "Percentage of Players": gender_dem_perc})

#Round numbers on DataFrame to 2
gender_dem_df = gender_dem_df.round(2)

#Display DataFrame
gender_dem_df
```

PURCHASING ANALYSIS (GENDER)

Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender


Create a summary data frame to hold the results


Optional: give the displayed data cleaner formatting


Display the summary data frame

```python
#Calculations for Purchasing Analysis by Gender
Total_Gender_Purchases = purchase_data_DF.groupby(["Gender"]).sum()["Price"].rename("Total Purchase Value")

Avg_Gender = purchase_data_DF.groupby(["Gender"]).mean()["Price"].rename("Average Purchase Price")

Gender_Count = purchase_data_DF.groupby(["Gender"]).count()["Price"].rename("Purchase Count")

Avg_PPP = Total_Gender_Purchases / Gender_Count

#Create DataFrame
Gender_Purchasing_DF = pd.DataFrame({"Purchase Count": Gender_Count,
                                    "Average Purchase Price": Avg_Gender,
                                    "Total Purchase Value": Total_Gender_Purchases,
                                    "Average Total Purchases Per Person": Avg_PPP})

#Round DataFrame to 2 decimal points
Gender_Purchasing_DF = Gender_Purchasing_DF.round(2
                                                 )
#Display DataFrame
Gender_Purchasing_DF
```

AGE DEMOGRAPHICS

Establish bins for ages


Categorize the existing players using the age bins. Hint: use pd.cut()


Calculate the numbers and percentages by age group


Create a summary data frame to hold the results


Optional: round the percentage column to two decimal points


Display Age Demographics Table

```python
#Create bins and labels
bins = ([0,10,15,20,25,30,35,40,100])
group_bins = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "Over 40"]

#Groupby and adding bins to new DF
bins_DF = purchase_data_DF.copy()
bins_DF["Age Groups"] = pd.cut(bins_DF["Age"], bins, labels=group_bins)
group_bins_DF = bins_DF.groupby(["Age Groups"])

#Calculations
bin_count = group_bins_DF["SN"].count()
total_count = purchase_data_DF["SN"].count()
bin_perc = (bin_count / total_count) * 100

#Create Data Frame
Age_Bins_DF = pd.DataFrame({"Percentage of Players": bin_perc,
                           "Total Count": bin_count})

#Round 2 decimal points
Age_Bins_DF = Age_Bins_DF.round(2)

#Display DF
Age_Bins_DF
```

PURCHASING ANALYSIS (AGE)

Bin the purchase_data data frame by age


Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


Create a summary data frame to hold the results


Optional: give the displayed data cleaner formatting


Display the summary data frame

```python
#Add bins to new DF and groupby
Age_Bins_DF2 = purchase_data_DF.copy()
Age_Bins_DF2["Age"] = pd.cut(purchase_data_DF["Age"], bins, labels=group_bins)
Age_Bins_DF3 = Age_Bins_DF2[["Price", "Age"]].groupby(["Age"]).count()
Age_Bins_DF3.rename(columns={"Price": "Purchase Count"}, inplace = True)

#Data Calculatins
Age_Bins_DF3["Average Purchase Price"] = Age_Bins_DF2[["Price", "Age"]].groupby(["Age"]).mean()
Age_Bins_DF3["Total Purchase Price"] = Age_Bins_DF2[["Price", "Age"]].groupby(["Age"]).sum()
Age_Bins_DF3["Avg Total Purchase per Person"] = (Age_Bins_DF3["Total Purchase Price"] / Age_Bins_DF["Total Count"])

#Round to 2 decimal points
Age_Bins_DF3.round(2)
```

TOP SPENDERS

Run basic calculations to obtain the results in the table below


Create a summary data frame to hold the results


Sort the total purchase value column in descending order


Optional: give the displayed data cleaner formatting


Display a preview of the summary data frame

```python
#Create new DataFrame
Top_Spender_DF = purchase_data_DF["Item ID"].groupby(purchase_data_DF["SN"]).count()
Top_Spender_DF = pd.DataFrame(data=Top_Spender_DF)
Top_Spender_DF.columns = ["Purchase Count"]

#Calculations
Top_Spender_DF["Average Purchase Price"] = purchase_data_DF["Price"].groupby(purchase_data_DF["SN"]).mean()
Top_Spender_DF["Total Purchase Value"] = purchase_data_DF["Price"].groupby(purchase_data_DF["SN"]).sum()

#Set to descending order
Top_Spender_DF.sort_values(by = ["Total Purchase Value"], ascending = False, inplace = True)

#Round to 2 decimal points
Top_Spender_DF.round(2)

#Display DF
Top_Spender_DF
```

MOST POPULAR ITEMS

Retrieve the Item ID, Item Name, and Item Price columns


Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


Create a summary data frame to hold the results


Sort the purchase count column in descending order


Optional: give the displayed data cleaner formatting


Display a preview of the summary data frame

```python
#Create New DF, combine functions
Pop_Items_DF = purchase_data_DF.groupby(["Item ID", "Item Name", "Price"])["Price"].agg(["count", "sum"])

#Rename Columns
Pop_Items_DF.columns = ["Purchase Count", "Total Purchase Value"]

#Reset and Set new Index
Pop_Items_DF.reset_index(inplace = True)
Pop_Items_DF.set_index(["Item ID", "Item Name"], inplace = True)

Pop_Items_DF = Pop_Items_DF[["Purchase Count", "Price", "Total Purchase Value"]]

#Sort in Descending order
Most_Pop_Items_DF = Pop_Items_DF.sort_values(by = "Purchase Count", ascending = False)

#Display top 10 in DF
Most_Pop_Items_DF.head(10)
```
